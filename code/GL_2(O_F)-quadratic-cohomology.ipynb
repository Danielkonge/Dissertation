```python
from main_functions import *
from numba import njit
```

```python
g_gl2_F_1 = np.array([1,2,7,8], dtype=np.int32) #1-graded (1,2,7,8)
g_gl2_F_2 = np.array([3,4,5,6], dtype=np.int32) #2-graded (3,4,5,6)
g_gl2_F = np.array([g_gl2_F_1,g_gl2_F_2], dtype=object); 
g_gl2_F
```

```python
# g = g^1 + g^2, we order them g^1, g^2,
# where g^1 = span(e_1, e_2, e_6, e_7), g^2 = span(e_3, e_4, e_5),
# so the prefered order is: 
# e_1 (1st), e_2 (2nd), e_7 (3rd), e_8 (4th),
# e_3 (5th), e_4 (6th), e_5 (7th), e_6 (8th)
# i.e., in a zero indexed list:
# (e_1 =) 1-->0, (e_2 =) 2-->1, 3-->4, 4-->5, 5-->6, 6-->7, 7-->2, 8-->3
e_sort_gl2_F = np.array([0,1,4,5,6,7,2,3], dtype=np.int32)
```

```python
@njit
def commutator_unram(a, b):
    sign = 1
    if a > b:
        a,b = b,a
        sign = -1
    # to make it easier later, we return lists with the answers, 
    # where a x+y = [x,y]
    if a==1 and b==7:
        return [(-sign,3)]
    elif a==1 and b==8:
        return [(-sign,4)]
    elif a==2 and b==7:
        return [(-sign,4)]
    elif a==2 and b==8:
        return [(sign,3)]
    else:
        return [(0,0)]
    
@njit
def commutator_ram(a, b):
    sign = 1
    if a > b:
        a,b = b,a
        sign = -1
    # to make it easier later, we return lists with the answers, 
    # where a x+y = [x,y]
    if a==1 and b==3:
        return [(2*sign,3)]
    elif a==1 and b==7:
        return [(-sign,2),(-sign,3),(sign,8)]
    elif a==1 and b==8:
        return [(-sign,4)]
    elif a==2 and b==7:
        return [(-sign,4)]
    elif a==3 and b==7:
        return [(2*sign,8)]
    else:
        return [(0,0)]
```

```python
# We call commutator so it compiles. 
commutator_unram(1,7), commutator_ram(1,7)
```
