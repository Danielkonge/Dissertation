```python
from main_functions import *
from numba import njit
```

```python
g_gl3_1 = np.array([1,7,8], dtype=np.int32) #1-graded (1,7,8)
g_gl3_2 = np.array([2,3,9], dtype=np.int32) #2-graded (2,3,9)
g_gl3_3 = np.array([4,5,6], dtype=np.int32) #3-graded (4,5,6)
g_gl3 = np.array([g_gl3_1,g_gl3_2,g_gl3_3], dtype=object); 
g_gl3
```

```python
# g = g^1 + g^2 + g^3, we order them g^1, g^2, g^3, 
# where g^1 = span(e_1, e_7, e_8), g^2 = span(e_2, e_3, e_8), g^3 = span(e_4, e_5, e_6)
# so the prefered order is: 
# e_1 (1st), e_7 (2nd), e_8 (3rd), 
# e_2 (4th), e_3 (5th), e_9 (6th), 
# e_4 (7th), e_5 (8th), e_6 (9th)
# i.e., in a zero indexed list:
# (e_1 =) 1-->0, (e_2 =) 2-->3, 3-->4, 4-->6, 5-->7, 6-->8, 7-->1, 8-->2, 9-->5
e_sort_gl3 = np.array([0,3,4,6,7,8,1,2,5], dtype=np.int32)
```

```python
@njit
def commutator(a, b):
    sign = 1
    if a > b:
        a,b = b,a
        sign = -1
    # to make it easier later, we return lists with the answers, where a x+y = [x,y]
    if a==1 and b==7:
        return [(-sign,2)]
    elif a==1 and b==8:
        return [(sign,3)]
    elif a==1 and b==9:
        return [(-sign,5),(-sign,4)] # for the sake of later code, we order these in reverse order
    elif a==2 and b==8:
        return [(-sign,4)]
    elif a==3 and b==7:
        return [(-sign,5)]
    elif a==7 and b==8:
        return [(-sign,9)]
    else:
        return [(0,0)]
```

```python
# We call commutator so it compiles. 
commutator(1, 7)
```
