```python
cohomology_dimensions_sl3 = np.zeros((16,24), dtype=np.int32)
cohomology_dimensions_sl3[0][0] = 1
cohomology_dimensions_sl3[1][2] = 3
cohomology_dimensions_sl3[3][5] = 6
cohomology_dimensions_sl3[4][6] = 3
cohomology_dimensions_sl3[4][7] = 3
cohomology_dimensions_sl3[5][8] = 6
cohomology_dimensions_sl3[6][9] = 7
cohomology_dimensions_sl3[7][11] = 9
cohomology_dimensions_sl3[8][12] = 9


for j in range(8):
    for t in range(12):
        if (j-15)+(23-t) > 4: # -(36-j) + (51-t) > 7 (part of H^n, n>7)
            cohomology_dimensions_sl3[15-j][23-t] = cohomology_dimensions_sl3[j][t]
```

```python
for j in range(16):
    for t in range(24):
        if cohomology_dimensions_sl3[j][t] != 0:
            print(f"dim H^({-j},{t}) = {cohomology_dimensions_sl3[j][t]}")
```

```python
graph_sl3 = nx.DiGraph()
for j1 in range(1,15):
    for t1 in range(1,23):
        if cohomology_dimensions_sl3[j1][t1] != 0:
            # graph_sl3.add_edge("(0,0)",f"({j1},{t1})")
            for j2 in range(j1,16-j1):
                for t2 in range(t1,24-t1):
                    if cohomology_dimensions_sl3[j2][t2] != 0:
                        if cohomology_dimensions_sl3[j1+j2][t1+t2] != 0:
                            graph_sl3.add_edge(f"({j1},{t1})",f"({j1+j2},{t1+t2})")
                            graph_sl3.add_edge(f"({j2},{t2})",f"({j1+j2},{t1+t2})")
```
